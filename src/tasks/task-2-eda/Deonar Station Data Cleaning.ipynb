```python
deonar_dataset = deonar_dataset.drop('Unnamed: 0', axis = 1)
```

Converting None to NaN val

```python
deonar_dataset= deonar_dataset.replace('None', np.nan)

# Converting type

deonar_dataset = deonar_dataset.astype({'PM2.5': 'float','PM10': 'float'})
```

```python
deonar_dataset.head()
```

```python
deonar_dataset['Unnamed: 24'].isnull().sum()
```

Unnamed 24 contains all Null values We should drop this!

```python
deonar_dataset = deonar_dataset.drop('Unnamed: 24', axis = 1)
```

```python
deonar_dataset.columns
```

Converting all columns to Float

```python
deonar_dataset = deonar_dataset.astype({'NO': 'float', 'NO2' : 'float','NOx' : 'float', 'NH3' : 'float',
       'SO2': 'float', 'CO': 'float', 'Ozone': 'float', 'Benzene': 'float', 'Toluene' : 'float',
       'Eth-Benzene': 'float', 'MP-Xylene' : 'float','RH' : 'float', 'WS' : 'float',
        'WD' : 'float', 'BP' : 'float', 'Xylene': 'float', 'AT': 'float', 'RF': 'float'})
```

```python
deonar_dataset.describe().T
```

Filling NaN Values

```python
columns = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3',
       'SO2', 'CO', 'Ozone', 'Benzene', 'Toluene', 'Eth-Benzene', 'MP-Xylene',
       'RH', 'WS', 'WD', 'BP', 'Xylene', 'AT','RF']

for col in columns:
    deonar_dataset[col] = deonar_dataset[col].replace(np.nan, deonar_dataset[col].median())
deonar_dataset.head()
```
