### Statistical tests

1. ANOVA tests for multiple samples

H0: All means are the same

Ha: All means are not the same

```python
from scipy.stats import f_oneway

statistic = f_oneway(powai_df['PM2.5'][:30],
                     colaba_df['PM2.5'][:30],
                     bkc_df['PM2.5'][:30]
                     )

statistic
```

- Here, pvalue is less than 0.05 so we will reject the null hypothesis and accept the alt hypothesis.

```python
powai_df['PM2.5'] = [float(str(i).replace("None", '0')) for i in powai_df['PM2.5']]
powai_df['PM2.5'].plot()
```

```python
# convert the datatype of each df
def dtype_convert(df):
    cols = ['PM2.5','PM10','NO','NO2','NOx','NH3','SO2','CO','Ozone']
    for col in cols:
        df[col] = [float(str(i).replace("None", '0')) for i in df[col]]
        
    df['From Date'] = pd.to_datetime(df['From Date'])
    df['To Date'] = pd.to_datetime(df['To Date'])
    return df
```

- Modeling individually pollutants

```python
new_po_df = np_df[['From Date','PM2.5','PM10','NO','NO2','NOx','NH3','SO2','CO','Ozone']]
new_po_df['Date'] = new_po_df['From Date'].dt.date
```

```python
# reaplce zero values to mean of the columns
new_po_df['PM2.5'] = new_po_df['PM2.5'].replace(0.0, new_po_df['PM2.5'].mean())
new_po_df['PM10'] = new_po_df['PM10'].replace(0.0, new_po_df['PM10'].mean())
new_po_df['NO'] = new_po_df['NO'].replace(0.0, new_po_df['NO'].mean())
new_po_df['NO2'] = new_po_df['NO2'].replace(0.0, new_po_df['NO2'].mean())
```

```python
new_po_df.info()
```

```python
new_po_df['PM10'].plot()
```

```python
new_po_df['PM10'] = new_po_df['PM10'].map(lambda x: new_po_df['PM10'].mean() if x > 600 else x)
new_po_df['PM10'].plot()
```

```python
# heatmap of the 
plt.figure(figsize=(7,7))
sns.heatmap(new_po_df.corr(), vmin=-1, vmax=+1)
plt.show()
```

```python
# pairplot between all the numerical columns
sns.pairplot(new_po_df)
```

```python
# outlier treatment
def outlier_func(df):
    df['PM2.5'] = df['PM2.5'].map(lambda x: df['PM2.5'].mean() if x > 600 else x)
    df['NO'] = df['NO'].map(lambda x: df['NO'].mean() if x > 250 else x)
    df['NO2'] = df['NO2'].map(lambda x: df['NO2'].mean() if x > 150 else x)
    df['NOx'] = df['NOx'].map(lambda x: df['NOx'].mean() if x > 300 else x)
    df['NH3'] = df['NH3'].map(lambda x: df['NH3'].mean() if x > 350 else x)
    df['SO2'] = df['SO2'].map(lambda x: df['SO2'].mean() if x > 100 else x)
    return df
```

```python
converted_po_df = outlier_func(new_po_df)
```

```python
converted_po_df['Date'] = pd.to_datetime(converted_po_df['Date'])
```

```python
converted_po_df.to_csv("final_powai_df.csv")
```
