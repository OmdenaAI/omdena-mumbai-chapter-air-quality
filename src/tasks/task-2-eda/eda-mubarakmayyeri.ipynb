```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Loading datasets
stations = os.listdir('downloads/')
df_list = []

for filename in stations:
    if filename.endswith('.csv'):
        # Load the CSV file into a dataframe
        df = pd.read_csv(os.path.join('Downloads', filename))
        
        # Create a variable with the same name as the CSV file
        variable_name = os.path.splitext(filename)[0]
        globals()[variable_name] = df
        
        # Saving the df names into list
        df_list.append(variable_name)
```

### Data Dictionary

```python
# Counting how station dataset contains each parameters
cols = {}
for station in df_list:
    for col in globals()[station].columns:
        if col not in cols:
            cols[col] = 1
        else:
            cols[col] += 1       
cols
```

```python
# Creating a data dictionary with abbrv and parameter names
parameters = ['Start Date', 'End Date', 'Particulate Matter less than 2.5 micron size', 'Particulate Matter less than 10 Micron size',
              'Nitric Oxide', 'Nitrogen dioxied', 'Oxides of Nitrogen', 'Ammonia', 'Sulphur Dioxide','Carbon Monoxide',
              'Ozone', 'Benzen', 'Toluene', 'Ethyl Benzene', 'M+P Xylene', 'Relative Humidity', 'Wind Speed', 'Wind Direcction',
              'Solar Radiation', 'Barometric Pressure', 'Xylene', 'Ambient Temperature', 'Rain Fall', 'TOT-RF',
              'Rack Temperature', 'Vertical Wind Speed']

param_abbv = list(cols.keys())

data_dictionary = dict()

for i in range(len(parameters)):
    data_dictionary[param_abbv[i]] = parameters[i]
    
data_dictionary
```

```python
params_df = pd.DataFrame(data_dictionary.items(), columns=['Abbrevations', 'Parametres'])
params_df.index = range(len(params_df))
```

```python
params_df['How many stations'] = [cols[abb] for abb in param_abbv]
```

```python
# Replacing the None values with NaN and converting dtype to float
for station in df_list:
    df = globals()[station]
    df = df.replace('None', np.NaN)
    
    # Converting dtype of data colums
    df[df.columns[:2]] = df[df.columns[:2]].apply(pd.to_datetime)
    df[df.columns[2:]] = df[df.columns[2:]].astype(float)
    globals()[station] = df
```
