### Statistical tests

1. ANOVA tests for multiple samples

H0: All means are the same

Ha: All means are not the same

```python
from scipy.stats import f_oneway

statistic = f_oneway(powai_df['PM2.5'][:30],
                     colaba_df['PM2.5'][:30],
                     bkc_df['PM2.5'][:30]
                     )

statistic
```

- Here, pvalue is less than 0.05 so we will reject the null hypothesis and accept the alt hypothesis.

```python
powai_df['PM2.5'] = [float(str(i).replace("None", '0')) for i in powai_df['PM2.5']]
powai_df['PM2.5'].plot()
```

```python
# convert the datatype of each df
def dtype_convert(df):
    cols = ['PM2.5','PM10','NO','NO2','NOx','NH3','SO2','CO','Ozone']
    for col in cols:
        df[col] = [float(str(i).replace("None", '0')) for i in df[col]]
        
    df['From Date'] = pd.to_datetime(df['From Date'])
    df['To Date'] = pd.to_datetime(df['To Date'])
    return df
```

```python
np_df = dtype_convert(powai_df)
np_df.info()
```

- Modeling individually pollutants

```python
new_po_df = np_df[['From Date','PM2.5','PM10','NO','NO2','NOx','NH3','SO2','CO','Ozone']]
new_po_df['Date'] = new_po_df['From Date'].dt.date
```

```python

new_po_df['PM2.5'] = new_po_df['PM2.5'].replace(0.0, new_po_df['PM2.5'].mean())
new_po_df['PM10'] = new_po_df['PM10'].replace(0.0, new_po_df['PM10'].mean())
new_po_df['NO'] = new_po_df['NO'].replace(0.0, new_po_df['NO'].mean())
new_po_df['NO2'] = new_po_df['NO2'].replace(0.0, new_po_df['NO2'].mean())
```
