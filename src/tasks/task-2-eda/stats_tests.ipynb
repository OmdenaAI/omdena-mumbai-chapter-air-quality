### Statistical tests

1. ANOVA tests for multiple samples

H0: All means are the same

Ha: All means are not the same

```python
from scipy.stats import f_oneway

statistic = f_oneway(powai_df['PM2.5'][:30],
                     colaba_df['PM2.5'][:30],
                     bkc_df['PM2.5'][:30]
                     )

statistic
```

- Here, pvalue is less than 0.05 so we will reject the null hypothesis and accept the alt hypothesis.
