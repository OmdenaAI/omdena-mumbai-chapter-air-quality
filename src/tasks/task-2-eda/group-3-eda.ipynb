### Reading Semi Cleaned Datasets

```python
# Loading datasets
stations = os.listdir('downloads/semi_cleaned_files/group_3')
df_list = []

for filename in stations:
    if filename.endswith('.csv'):
        # Load the CSV file into a dataframe
        df = pd.read_csv(os.path.join('downloads/semi_cleaned_files/group_3', filename), index_col=0)
        
        # Create a variable with the same name as the CSV file
        variable_name = os.path.splitext(filename)[0]
        globals()[variable_name] = df
        
        # Saving the df names into list
        df_list.append(variable_name)
```
