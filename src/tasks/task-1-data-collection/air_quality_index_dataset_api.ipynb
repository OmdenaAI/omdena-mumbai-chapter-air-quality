### We can use ozon3 library for AQI data collection
Library Github : https://github.com/Ozon3Org/Ozon3 \

I am going to show how we can use OZONE library to collect air quality data. 
Ozone is a python wrapper package to free API services hosted by [src: https://aqicn.org]. It handles the process of interacting with the API sever, so instead of wrangling bare JSONs, it returns you a pandas DataFrame.

The API service requires you to register for an API token for access.

### Import essential library

```python
import pandas as pd
import ozon3 as ooo ### importing ozone based library
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
# Define token
access_token = "8638018165895e421b7eab4dc8baf70a6b182063"
# Instantiate ozone object
o3 = ooo.Ozon3(access_token)
```

### Getting live data from the city we use get_city_air method

```python
# return single row
o3.get_city_air("Mumbai")
```

### For getting live data from multiple city we can use get_multiple_city_air method

```python
o3.get_multiple_city_air(
    ['New Delhi', 'Mumbai', 'Chennai', 'New York', 'Paris'])
```

### We can get live data using coordinates. It returns the nearest station coordinates data.

```python
o3.get_coordinate_air(19.00000, 73.000000)
```

### We can use bounding box to get all live data from different stations from India

```python
LEFT, BOTTOM, RIGHT, TOP = 68.1766451354, 7.96553477623, 97.4025614766, 35.4940095078

df = o3.get_range_coordinates_air((TOP, LEFT), (BOTTOM, RIGHT))
df
```

### Above dataframe contains data from the stations in india. We can try to get data only from stations in Mumbai

```python
df_s = df[df['station'].notnull()]
df_s[df_s['station'].str.contains('Mumbai')]
```

### Ozon3 api also forcast 7 day forecasts for some parameters.

```python
o3.get_city_forecast("Mumbai")[:10]
```

### OZONE also provides way to collect historicaldata but it is from 2014 onwards but it is slow.

```python
o3.get_historical_data('Mumbai')[:10]
```
