```python
api_key = '8988e125fc03d86da0c21c3afdb19258'

#mumbai coordinates
lat = '19.076090'
lon = '72.877426'

#time frame 
start = '1575072000'
end = '1678054634'

url = 'http://api.openweathermap.org/data/2.5/air_pollution/history?lat={}&lon={}&start={}&end={}&appid={}'.format(lat, lon, start, end, api_key)
```

```python
response = requests.request('GET', url)
```

```python
json_response = response.json()
```

```python
json_response
```

```python
aqi = []
co = []
no = []
no2 = []
o3 = []
so2 = []
pm2_5 = []
pm10 = []
nh3 = []
dt = []

for data in json_response['list']:
    aqi.append(data['main']['aqi'])
    co.append(data['components']['co'])
    no.append(data['components']['no'])
    no2.append(data['components']['no2'])
    o3.append(data['components']['o3'])
    so2.append(data['components']['so2'])
    pm2_5.append(data['components']['pm2_5'])
    pm10.append(data['components']['pm10'])
    nh3.append(data['components']['nh3'])
    dt.append(data['dt'])
```

```python
data = {
    'date_time': dt,
    'aqi': aqi,
    'co': co,
    'no': no,
    'no2': no2,
    'o3': o3,
    'so2': so2,
    'pm2_5': pm2_5,
    'pm10': pm10,
    'nh3': nh3 
}
```

```python
df = pd.DataFrame.from_dict(data)

df.head()
```

```python
df.to_csv('data.csv')
```

```python
def retrieve_data(place, coord):
    a, b = coord
    q = str(a) + ',' + str(b) 

    try:
        test_url = 'http://api.worldweatheronline.com/premium/v1/weather.ashx?key={}&q={}&num_of_days={}&tp=24&format=json&aqi=yes'.format(api_key, q, str(num_days))

        response = requests.request('GET', test_url)

        test_res = response.json()
    except:
        print('Request failed')
        
    date = []
    co = []
    no2 = []
    o3 = []
    so2 = []
    pm2_5 = []
    pm10 = []
    minTemp = []
    maxTemp = []
    avgTemp = []
    windSpeed = []
    windDirection = []
    humidity = []
    visibility = []
    pressure = []
    heatIndex = []
    dewPoint = []
    windChill = []
    windgust = []

    for data in test_res['data']['weather']:
        date.append(data['date'])
        if len(data['air_quality']) != 0:
            co.append(data['air_quality']['co'])
            no2.append(data['air_quality']['no2'])
            o3.append(data['air_quality']['o3'])
            so2.append(data['air_quality']['so2'])
            pm2_5.append(data['air_quality']['pm2_5'])
            pm10.append(data['air_quality']['pm10'])
        else:
            co.append(None)
            no2.append(None)
            o3.append(None)
            so2.append(None)
            pm2_5.append(None)
            pm10.append(None) 
        minTemp.append(data['mintempC'])
        maxTemp.append(data['maxtempC'])
        avgTemp.append(data['avgtempC'])
        windSpeed.append(data['hourly'][0]['windspeedKmph'])
        windDirection.append(data['hourly'][0]['winddirDegree'])
        humidity.append(data['hourly'][0]['humidity'])
        visibility.append(data['hourly'][0]['visibility'])
        pressure.append(data['hourly'][0]['pressure'])
        heatIndex.append(data['hourly'][0]['HeatIndexC'])
        dewPoint.append(data['hourly'][0]['DewPointC'])
        windChill.append(data['hourly'][0]['WindChillC'])
        windgust.append(data['hourly'][0]['WindGustKmph'])

    data = {
        'date': date,
        'co': co,
        'no2': no2,
        'o3': o3,
        'so2': so2,
        'pm2_5': pm2_5,
        'pm10': pm10,
        'minTempC': minTemp,
        'maxTempC': maxTemp,
        'avgTempC': avgTemp,
        'windSpeed': windSpeed,
        'windDirection': windDirection,
        'humidity': humidity,
        'visibility': visibility,
        'pressure': pressure,
        'heatIndex': heatIndex,
        'dewPoint': dewPoint,
        'windChill': windChill,
        'windGust': windgust
    }

    df = pd.DataFrame.from_dict(data)

    df.to_csv(place+'.csv') 
```
